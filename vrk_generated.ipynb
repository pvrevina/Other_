```python
import numpy as np
import re
```

```python
titles = open('titles.txt').read()
```

```python
abstracts = open('abstracts.txt', encoding = 'utf-8').read()
```

```python
txt = titles + ' ' + abstracts
len(txt)
```

```python
txt = re.sub('[А-Я]\d*\:\s', '', txt)
txt = re.sub('(\.\(\-|\(\-|г\.)', '', txt)
txt = re.sub('(\'|“|„|’|”|«|»)', '"', txt)
txt = re.sub('(…|\,\.)', '.', txt)
txt = re.sub('(>|])', ')', txt)
txt = re.sub('(<|\[)', '(', txt)
```

```python
corpus = txt.split()
```

```python
def make_pairs(corpus):
    for i in range(len(corpus)-1):
        yield (corpus[i], corpus[i+1])
```

```python
def make_triples(corpus):
    for i in range(len(corpus)-2):
        yield (corpus[i], corpus[i+1], corpus[i+2])
```

```python
pairs = make_pairs(corpus)
triples = make_triples(corpus)
```

```python
pair_dict = {}
for word_1, word_2 in pairs:
    if word_1 in pair_dict.keys():
        pair_dict[word_1].append(word_2)
    else:
        pair_dict[word_1] = [word_2]
```

```python
triple_dict = {}
for word_1, word_2, word_3 in triples:
    if str(word_1 + ' ' + word_2) in triple_dict.keys():
        triple_dict[str(word_1 + ' ' + word_2)].append(word_3)
    else:
        triple_dict[str(word_1 + ' ' + word_2)] = [word_3]
```

```python
def generate(corpus, pair_dict, triple_dict, length=30, tr_pr=0.3):
    
    first_word = np.random.choice(corpus)
    while first_word.islower() or len(first_word) <= 2 or first_word not in pair_dict.keys() or first_word[-1] == '.' or first_word[-1] == ')' or first_word[-1] == '"' or first_word[0] == '(':
        first_word = np.random.choice(corpus)
    sec_word = np.random.choice(pair_dict[first_word])
    chain = str(first_word + ' ' + sec_word)
    n_words = length
    
    i = 0
    
        
    while i <= n_words or chain[-1] != '.':
        
        rnd = np.random.rand()
        if rnd < tr_pr:
            
            next_word = np.random.choice(triple_dict[' '.join(re.sub('\d+', '', chain).split()[-2:])])
            next_word = re.sub('\%', str(int(100 * np.random.rand())) + '%', next_word)
            next_word = re.sub('тыс\.', str(int(100 * np.random.rand())) + 'тыс.', next_word)
            next_word = re.sub('\$', str(int(100 * np.random.rand())) + '$', next_word)
            next_word = re.sub('-лет', str(int(100 * np.random.rand())) + '-лет', next_word)
            
            chain += (' ' + next_word)
            i += 1
            
        else:
            next_word = np.random.choice(pair_dict[' '.join(re.sub('\d+', '', chain).split()[-1:])])
            next_word = re.sub('\%', str(int(100 * np.random.rand())) + '%', next_word)
            next_word = re.sub('тыс\.', str(int(100 * np.random.rand())) + 'тыс.', next_word)
            next_word = re.sub('\$', str(int(100 * np.random.rand())) + '$', next_word)
            next_word = re.sub('-лет', str(int(100 * np.random.rand())) + '-лет', next_word)
            
            chain += (' ' + next_word)
            i += 1
    
    return chain
```

```python
for i in range(10):
    
    print(generate(corpus, pair_dict, triple_dict, length=20, tr_pr=0.9))
    print()
    print()
```
